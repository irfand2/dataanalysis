```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import statistics as s
import math

data_arcelor =  pd.read_csv(r'C:\Users\Atilla\Downloads\MT.AS.csv')
data_arcelor = data_arcelor.dropna(axis=0)
arcelor_close = data_arcelor["Close"]
data_airbus = pd.read_csv(r'C:\Users\Atilla\Downloads\AIR.PA.csv')
data_airbus = data_airbus.dropna(axis=0)
data_airbus = data_airbus.drop(255, axis=0)
```

```python
airbus_close = data_airbus["Close"]
arcelor_std = (arcelor_close - arcelor_close.mean()) / arcelor_close.std(ddof=0)
predict_airbus_std = arcelor_std * 0.9066156
predict_airbus = predict_airbus_std * airbus_close.std(ddof=0) + airbus_close.mean()
```
